# Creation de Games dataframe 

Generation de liste unique 

```python
game_id_list = list(range(0, len(all_data)))
game_id_list
```

Generer les liste ids du genre dans le dataframe 

```python
genre_id = []
for index, row in all_data.iterrows():
    genre_id.append((genre_data.loc[genre_data['genre'] == row[2]]['genre_id']).values[0])
genre_id
```

Genererer les liste console ids dans le dataframe 

```python
console_id = []
for index, row in all_data.iterrows():
    console_id.append((console_data.loc[console_data['console'] == row[1]]['console_id']).values[0])
console_id
```

Genererer les liste publisher ids dans le dataframe 

```python
publisher_id = []
for index, row in all_data.iterrows():
    publisher_id.append((publisher_data.loc[publisher_data['publisher'] == row[3]]['publisher_id']).values[0])
publisher_id
```

Genererer les liste devellopeur ids dans le dataframe 

```python
developer_id = []
for index, row in all_data.iterrows():
    if str(row[4]) == 'nan':
        developer_id.append(pd.NA)
    else:
        developer_id.append((developer_data.loc[developer_data['developer'] == row[4]]['developer_id']).values[0])
developer_id
```

```python
all_data.columns
```

```python
games_data = pd.DataFrame({
    'game_id' : game_id_list,
    'title' : all_data['title'],
    'annees' : all_data['annees'],
    'release_date' : all_data['release_date'],
    'critic_score' : all_data['critic_score'],
    'genre_id' : genre_id,
    'console_id' : console_id,
    'publisher_id' : publisher_id,
    'developer_id' : developer_id
})
games_data
```

```python
games_data.dtypes
```

```python
games_data["release_date"] = games_data["release_date"].astype('datetime64[ns]')
games_data["developer_id"] = games_data["developer_id"].astype("Int64")
```
